```python
def sum_of_diagonal(data):
    """
    Функция возвращает сумму элементов на диоганали квадратной матрицы.
    если входные аданные не являются квадратной матрицой функция вернет 'Error
    """
    
    sum_of_elements = 0
    
    if check_data(data):
        for i in range(len(data)):
            sum_of_elements += data[i][i]
    else:
        return 'Error'
        
        
    return sum_of_elements

def check_data(data):
    """
    Проверка является ли функция квадратной матрицей
    """
    
    for line in data:
        if len(line) != len(data):
            return False
        
    return True
```

```python
def sum_of_squares(data):
    """
    Функция возвращает сумму квадратов всех элементов которые могут быть числами
    """
    
    sum_of_squares = 0

    for element in data:
        
        if type(element) == int:
            sum_of_squares += element**2
            
        if type(element) == float:
            sum_of_squares += element**2
            
        if type(element) == str:
            if str.isnumeric(element):
                sum_of_squares += float(element)**2
                
    return sum_of_squares
```

```python
def maximum_exchange():

    r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')

    valute = r.json()['Valute']

    max_value = 0
    max_name = ''

    for element in valute:
        if valute[element]['Value'] > max_value:
            max_value = valute[element]['Value']
            max_name = element
        
    return max_name
```

```python
class Rate:
    def __init__(self, format='value'):
        self.format = format
    
    def exchange_rates(self):
        """
        Возвращает ответ сервиса с информацией о валютах в виде:
        
        {
            'AMD': {
                'CharCode': 'AMD',
                'ID': 'R01060',
                'Name': 'Армянских драмов',
                'Nominal': 100,
                'NumCode': '051',
                'Previous': 14.103,
                'Value': 14.0879
                },
            ...
        }
        """
        r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return r.json()['Valute']
    
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate('full').make_format('EUR')
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        
        Rate('value').make_format('EUR')
        79.4966
        """
        response = self.exchange_rates()
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            if self.format == 'value':
                return response[currency]['Value']
        
        return 'Error'
    
    def eur(self, diff = False):
        """Возвращает курс евро на сегодня в формате self.format"""
        if diff == False:
            return self.make_format('EUR')
        if diff == True:
            self.format = 'full'
            response = self.make_format('EUR')
            return response['Value'] - response['Previous']
    
    def usd(self, diff = False):
        """Возвращает курс доллара на сегодня в формате self.format"""
        if diff == False:
            return self.make_format('USD')
        if diff == True:
            self.format = 'full'
            response = self.make_format('USD')
            return response['Value'] - response['Previous']
    
    def found_name(self, currency):
        """
        Возвращает название валюты currency:
        Rate().found_name('EUR')
        'Евро'
        """
        response = self.exchange_rates()
        
        if currency in response:
            return response[currency]['Name']
        
        return 'Error'
```

```python
def Fibonacci_numbers(number, list_of_numbers=[]):
    """
    Функция возвращает первые n чисел фибоначе
    """
            
    if len(list_of_numbers) > 1:
        list_of_numbers.append(list_of_numbers[-1] + list_of_numbers[-2])
    if len(list_of_numbers) == 1:
        list_of_numbers.append(1)
    if len(list_of_numbers) < 1:
        list_of_numbers = [1]
            
    if len(list_of_numbers) < number:
        Fibonacci_numbers(number, list_of_numbers)
        
    return (list_of_numbers)
```

```python
def list_to_dict(data_list, data_dict = {}):
    
    if len(data_dict) == 0:
        data_dict = {data_list[-2]: data_list[-1]}
        data_list.pop(-1)
        data_list.pop(-1)
        
    if len(data_list) != 0:
        data_dict = {data_list[-1]: data_dict}
        data_list.pop(-1)
        if len(data_list) != 0:
            list_to_dict(data_list, data_dict)
        else:
            return data_dict       
```
